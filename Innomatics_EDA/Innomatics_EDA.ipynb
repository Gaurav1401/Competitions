### Hypothesis Testing

```python
duo = df[df['matchType'] == 'duo']
```

**Your Friend is claiming that a player can be revived more than 3 times in a duo game. You what to check this using hypothesis testing**

Null Hypthesis: $$H_0 \Rightarrow \mu \leq 3$$

Alternate Hypthesis: $$H_1 \Rightarrow \mu > 3$$

```python
from scipy.stats import norm
```

```python
def z_score(sample_size, sample_mean, pop_mean, pop_std):
    numerator = sample_mean - pop_mean
    denomenator = pop_std / sample_size**0.5
    return numerator / denomenator
```

```python
np.std(duo['revives'])
```

```python
np.mean(duo['revives'])
```

```python
sample_size = duo.shape[0]
sample_mean = np.mean(duo['revives'])
pop_mean = 3
pop_std = np.std(duo['revives'])
```

```python
confidence_level = 0.95

alpha = 1 - confidence_level

z_critical = norm.ppf(1 - alpha) # Two tailed Z score for 95% Confidence Level

print(z_critical)
```

```python
# Calculating the z-score

z = z_score(sample_size, sample_mean, pop_mean, pop_std)

print(z)
```

```python
# Conclusion using z test

if(np.abs(z) > z_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

<font color = blue>**Regarding the current samples, we reject the null hypothesis**</font>

<font color = blue>**So, it is possible that a player can be revived more than 3 times in a game if he forms a duo**</font>
